## Introduccin

En este TP se pretende analizar dos esquemas de comunicacin: en uno se utilizan repetidores digitales
y en el otro se utilizan repetidores analgicos. En ambos casos se quiere transmitir smbolos que representan 1 bit.
Si el bit a enviar es $1$, el smbolo asociado es $X_1 = A$; si en cambio se quiere transmitir el bit $0$,
el smbolo es $X_1 = A$, donde $A > 0$. Es decir, la informacin transmitida se puede modelar como una variable
aleatoria discreta de soporte $\{A, A\}$. Asumiremos que ambos smbolos tienen probabilidad $1/2$.

Ambos sistemas de comunicaciones tienen $n$ etapas en cascada, es decir, hay $n  1$ repetidores.
En cada etapa, los smbolos se envan a travs de un canal de comunicaciones que puede ser modelado
por un factor de atenuacin $h$ y por la adicin de ruido $W_i$ , $i = 1, \dots , n$.
Asumiremos que la distribucin de este ruido es gaussiana de media nula y varianza $\sigma^{2}$ para cada canal,
es decir, $W_i \sim \mathcal{N}(0, \sigma^2)$, y que los ruidos de distintas etapas son independientes.
La diferencia entre ambos repetidores es la manera de procesar el smbolo recibido $Y_i$.
